```python
#1.
import numpy as np
x=np.random.random((10,1))
print(np.mean(x))
```

```python
#2.
x=np.random.random((20,1))
sd=np.std(x)
print('Standard deviation is:',sd)
v=np.var(x)
print('Variance is:',v)
```

```python
#3.
x1=np.random.random((10,20))
x2=np.random.random((20,25))
x3=np.matmul(x1,x2)
x4=np.sum(x3)
print('Sum of new formed array elements is',x4)
print('Matrix Multiplication of x1 and x2 gives',x3)
```

```python
#4.
a= np.random.rand(10,1)
print(a)

def func(x):
    return (1 / (1 + np.exp(-x)))
print('*'*10)
r = np.apply_along_axis(func, 0, a)
print(r)
```
